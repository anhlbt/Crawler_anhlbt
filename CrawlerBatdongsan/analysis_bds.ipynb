```python
import pandas as pd
from unidecode import unidecode
bds = pd.read_csv("./batdongsan_spider/batdongsan-2020-11-30.csv", sep='|')
```

```python
bds['province'] = bds['address'].apply(lambda x: unidecode(x.split(',')[-1].strip()))
bds['district'] = bds['address'].apply(lambda x: unidecode(x.split(',')[-2].strip()))
# bds['ward'] = bds['address'].apply(lambda x: unidecode(x.split(',')[-3].strip()))
```
